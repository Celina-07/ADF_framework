```python
import pandas as pd
import json 
import numpy as np
from tqdm import tqdm
import ast
import seaborn as sns
from matplotlib import pyplot as plt
```

# Import data

## Global results

```python
complete_results_baseline = pd.read_csv('baseline/complete_baseline.csv', index_col=0)
complete_results_ADF = pd.read_csv('ADF/ADF_complete.csv', index_col=0)
# complete_results_nopers = pd.read_csv('no_pers/complete_results.csv', index_col=0)
complete_results_greedy = pd.read_csv('greedy/complete_greedy.csv', index_col=0)
```

```python
complete_results_baseline['ILD'] = 1 - complete_results_baseline['ILS']
complete_results_ADF['ILD'] = 1 - complete_results_ADF['ILS']
# complete_results_nopers['ILD'] = 1 - complete_results_nopers['ILS']
complete_results_greedy['ILD'] = 1 - complete_results_greedy['ILS']
```

# Evolution of performances

```python
greedy = complete_results_greedy.reset_index(drop=True)
greedy.index = np.arange(0, 1.1, 0.1)

ADF = complete_results_ADF.reset_index(drop=True)
ADF.index = np.arange(0, 1.1, 0.1)
```

```python
baseline3_precision_x = [0]
baseline3_precision_y = [0.227]

fig, ax = plt.subplots(figsize=(10,8))
ax.axhline(y=0.224, linestyle='--', color='red', zorder=2)
ax.plot(greedy['Precision'], marker='s', color='dodgerblue')
ax.scatter(baseline3_precision_x, baseline3_precision_y, color='orangered', s=200, zorder=4)
ax.plot(ADF['Precision'], marker='s', color='darkorange', zorder=3)
# ax.axhline(y=0.224, linestyle='--', color='darkblue')
ax.legend(['A', 'AD', 'AF', 'ADF'], fontsize=24)
ax.set_xlabel(r'$\lambda$ - $\alpha$', fontsize=24)
ax.set_xticks(np.arange(0, 1.1, 0.1))

ax.set_ylabel('Precision@20', fontsize=24)
ax.grid()
```
